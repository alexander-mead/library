```python
# Add my own libraries to the path
import sys
sys.path.append('/Users/Mead/Physics/library/python')

# Make matplotlib work nicely
%matplotlib notebook
%matplotlib inline

# Import statements
import numpy as np
import matplotlib.pyplot as plt
import mead_statistics as statistics

# Distribution parameters
a = 1.
b = 1.

# Number to draw
n = 100000

# Probability distribution function (need not be normalised)
def p(x,a,b):
    return ((x/a)**b)*np.exp(-x/a)

# Set range in x, make arrays of x and C(x) (cumulative distribution) 
xmin = 0.
xmax = 10.
nx = 256
x = np.linspace(xmin,xmax,nx)
Cx = statistics.cumulative(x,p,xmin,a,b)

# Calculate normalisation (not necessary for drawing, only for plot)
norm = statistics.normalisation(p,xmin,xmax,a,b)
mean = statistics.moment(1,p,xmin,xmax,a,b)
var  = statistics.variance(p,xmin,xmax,a,b)
print('Probability normalised to:', norm)
print('Mean:', mean)
print('Variance:', var)
print()

# Draw random numbers from distribution
xi = []
for i in range(n):
    xi.append(statistics.draw_from_distribution(x,Cx))
    
# Plot draws
plt.plot(x,p(x,a,b)/norm);
plt.hist(xi,bins='auto',density='True');
```
