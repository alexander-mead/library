```python
# Add my own libraries to the path
import sys
sys.path.append('/Users/Mead/Physics/library/python')

# Make matplotlib work nicely
%matplotlib notebook
%matplotlib inline

# Import statements
import numpy as np
import matplotlib.pyplot as plt
import mead_statistics as statistics

# Distribution parameters
a = 1.
b = 1.

# Number to draw
n = 100000

# Probability distribution function (need not be normalised)
def p(x,a,b):
    return ((x/a)**b)*np.exp(-x/a)

# Set range in x, make arrays of x and C(x) (cumulative distribution) 
xmin = 0.
xmax = 10.
nx = 256
x = np.linspace(xmin,xmax,nx)
Cx = statistics.cumulative(x,p,xmin,a,b)

# Calculate normalisation (not necessary for drawing, only for plot)
norm = statistics.normalisation(p,xmin,xmax,a,b)
mean = statistics.moment(1,p,xmin,xmax,a,b)
var  = statistics.variance(p,xmin,xmax,a,b)
print('Probability normalised to:', norm)
print('Mean:', mean)
print('Variance:', var)
print()

# Draw random numbers from distribution
xi = []
for i in range(n):
    xi.append(statistics.draw_from_distribution(x,Cx))
    
# Plot draws
plt.plot(x,p(x,a,b)/norm);
plt.hist(xi,bins='auto',density='True');
```

```python
# Add my own libraries to the path
import sys
sys.path.append('/Users/Mead/Physics/library/python')

# Import statements
import numpy as np
import mead_statistics as statistics
from matplotlib import pyplot as plt

# Inline plots
%matplotlib inline

# Parameters
n = 1000000 # Number of random draws
m = 101 # Number of bins

# Define a normalised 2D Gaussian
def Gaussian(x,y):
    fx = np.exp(-x**2/2.)
    fy = np.exp(-y**2/2.)
    A = 1./(2.*np.pi)
    return A*fx*fy

# Set range for plot
rmin = -5. # Minimum x,y value in terms of sigma 
rmax = 5. # Maximum x,y value in terms of sigma 
x = np.linspace(rmin,rmax,m) # Create array of x values
dx = (rmax-rmin)/np.real(m) # Pixel size
y, dy = x, dx # y is the same as x
X,Y = np.meshgrid(x,y) # Necessary to make 2D plot
Z = Gaussian(X,Y) # Make Z array with Gaussian

# Plot
fig, ax = plt.subplots()
p = plt.pcolor(x-dx/2.,y-dy/2.,Z) # dx/2 necessary because otherwise x,y interpreted as pixel corners
plt.xlim((rmin,rmax))
plt.ylim((rmin,rmax))
plt.axis('scaled')
cb = fig.colorbar(p)
plt.show()

# Draw x and y independently
xs = np.random.normal(size=n)
ys = np.random.normal(size=n)

# Plot
fig, ax = plt.subplots()
plt.hist2d(xs,ys,bins=m,range=[[rmin,rmax],[rmin,rmax]],normed=True)
plt.axis('scaled')
plt.show()

# Use my draw from 2D function
xs,ys = statistics.draw_from_2D(n,Gaussian,rmin,rmax,m,rmin,rmax,m)

# Plot
fig, ax = plt.subplots()
plt.hist2d(xs,ys,bins=m,range=[[rmin,rmax],[rmin,rmax]],normed=True)
plt.axis('scaled')
plt.show()
```
