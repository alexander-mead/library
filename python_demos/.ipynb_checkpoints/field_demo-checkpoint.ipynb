```python
# Add my own libraries to the path
import sys
sys.path.append('/Users/Mead/Physics/library/python')

# Import libraries
import mead_field as field
import numpy as np
import matplotlib.pyplot as plt

power_norm = 10.
power_index = 2.
power_kdamp = 1.
field_mean = 0.
field_size = 10.
field_mesh = 256

plot_size = 8

def power_spectrum(k, A, n, kdamp):
        return A*(k**n)*np.exp(-k/kdamp)
    
field = field.make_Gaussian_random_field_2D(field_mean, 
                                               power_spectrum, 
                                               field_size, 
                                               field_mesh, 
                                               power_norm, 
                                               power_index, 
                                               power_kdamp)

plt.figure(figsize=(plot_size, plot_size))
image = plt.imshow(field, cmap=plt.cm.coolwarm, extent=[0.,field_size,0.,field_size])
```
